# 03 Data Cleaning

```python
df_clean=df_pokemons
df_clean.head(2)
```

```python
df_clean.columns=df_clean.columns.str.strip()
df_clean.head(2)
```

```python
df_clean.columns=df_clean.columns.str.lower()
df_clean.head(2)
```

```python
df_clean.columns=df_clean.columns.str.replace(' ','_')
df_clean.head(2)
```
