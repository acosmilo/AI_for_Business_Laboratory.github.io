# 03 Data Cleaning

```python
df_clean=df_pokemons
df_clean.head(2)
```

```python
df_clean.columns=df_clean.columns.str.strip()
df_clean.head(2)
```

```python
df_clean.columns=df_clean.columns.str.lower()
df_clean.head(2)
```

```python
df_clean.columns=df_clean.columns.str.replace(' ','_')
df_clean.head(2)
```

```python
df_clean["type_2"].isnull().sum()
```

```python
df_clean["type_2"].notna().sum()
```

```python
missing_table = df_clean.isna().sum().to_frame("missing_count")
missing_table
```

```python
df_clean["number_of_types"] = np.where(
    df_clean["type_2"].isna(),
    "Single Type",
    "Dual Type"
)
df_clean
```

```python
df_clean = df_clean.rename(columns={
    "legendary": "legendary_status"
})
df_clean.head(1)
```

```python
df_clean["legendary_status"] = np.where(
    df_clean["legendary_status"]==0,
    "Non Legendary",
    "Legendary"
)
df_clean
```

```python
df_clean["final_evolution"] = np.where(
    df_clean["final_evolution"]==0,
    "Non ultimate",
    "Ultimate"
)
df_clean
```

```python
duplicate_rows = df_clean.duplicated().sum()
duplicate_rows
```

```python
unique_values = df_clean.nunique(dropna=False)
unique_values
```
